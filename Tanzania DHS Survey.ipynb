## 1. Trends in Family planning and Fertility Rate

```python
# Data
years = [2016, 2010, 2004, 1999, 1996, 1991]
tfr = [5.2, 5.4, 5.7, 5.6, 5.8, 6.2]
contraception_use = [38.4, 23.6, 17.6, 15.6, 11.7, 5.9]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot Total Fertility Rate
ax.plot(years, tfr, marker='o', label='Total Fertility Rate', color='blue')

# Plot Current Use of Contraception
ax.plot(years, contraception_use, marker='o', label='Current Contraception Use', color='green')

# Add labels and title
ax.set_xlabel('Years')
ax.set_ylabel('Rate (%)')
ax.set_title('Total Fertility Rate vs. Current Contraception Use Over Time')

# Add a legend
ax.legend()

# Show the plot
plt.grid(True)
plt.show()

```

The first visualization presents a line chart that compares the Total Fertility Rate (TFR) and Current Use of a Modern Method of Contraception over the years 1991 to 2016.
The TFR measures the average number of children a woman is expected to have during her lifetime, while contraception use reflects the percentage of women using modern contraception methods.
Key Observations:

Total Fertility Rate (TFR): We observe a gradual decline in the TFR from 6.2 in 1991 to 5.2 in 2015. This decline suggests a reduction in the average number of children women are having, possibly indicating changing societal and economic factors.

Contraception Use: Current use of modern contraception methods shows significant variation. In 1991, only 5.9% of women used modern contraception, which increased to 38.4% by 2015. This points to an increasing awareness and accessibility of family planning services over time.

```python
import matplotlib.pyplot as plt

# Data
years = [2015, 2010, 2004, 1999, 1996, 1991]
tfr = [5.2, 5.4, 5.7, 5.6, 5.8, 6.2]
contraception_use = [38.4, 23.6, 17.6, 15.6, 11.7, 5.9]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Create a scatter plot
ax.scatter(contraception_use, tfr, c=years, s=100, alpha=0.8)

# Add labels and title
ax.set_xlabel('Current Use of Contraception (%)')
ax.set_ylabel('Total Fertility Rate')
ax.set_title('Relationship Between Total Fertility Rate and Contraception Use')

# Show the plot
plt.show()
```

When we observe the scatter plot, we can identify a general trend. As "Current Use of a Modern Method of Contraception" increases, there seems to be a corresponding decrease in the TFR. This suggests a potential inverse relationship between these two variables.
The scatter plot provides valuable insights into the complex relationship between family planning and fertility rates. While it's essential to recognize the general trend of decreasing fertility rates with increased contraception use, we must also acknowledge that numerous factors contribute to fertility rates, including cultural, social, economic, and policy influences.

## Unmet need for family planning overtime

```python
import matplotlib.pyplot as plt

# Data
years = [2015, 2010, 2004, 1999, 1996, 1991]
unmet_need = [22.1, 22.3, 24.3, 22.3, 26.0, 27.8]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar chart for Unmet Need for Family Planning
ax.bar(years, unmet_need,color='orange',alpha=0.7, align='center')

# Add labels and title
ax.set_xlabel('Years')
ax.set_ylabel('Unmet Need for Family Planning (%)')
ax.set_title('Unmet Need for Family Planning Over Time')

# Show the plot
plt.xticks(years)
plt.show()
```

Unmet Need for Family Planning: The bar chart highlights the percentage of women with an unmet need for family planning. Unmet need measures the proportion of women who wish to avoid pregnancy but are not using contraception.

Yearly Variation: The data shows fluctuations in unmet need, with a peak of 27.8% in 1991. However, there is a general trend of decline in unmet need, reaching 22.1% in 2015. This suggests progress in providing access to family planning services.

```python
import matplotlib.pyplot as plt

# Data
years = [2015, 2010, 2004, 1999, 1996, 1991]
unmet_need = [22.1, 22.3, 24.3, 22.3, 26.0, 27.8]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Create a box plot for Unmet Need for Family Planning
ax.boxplot(unmet_need, vert=False, labels=['Unmet Need'], patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Add labels and title
ax.set_xlabel('Percentage')
ax.set_title('Distribution of Unmet Need for Family Planning Over Time')

# Show the plot
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
